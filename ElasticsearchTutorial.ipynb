# I. Managing indices and mapppings

```python
# 1. initialize the client
from elasticsearch import Elasticsearch

HOST = "https://localhost:9200"
ELASTIC_USER = "elastic"
# The password for the 'elastic' user generated by Elasticsearch
ELASTIC_PASSWORD = "BD=khuw3Z2Sk40Ue-VwL"
# The path of ca certificates 
CA_CERTS = "/etc/elasticsearch/certs/http_ca.crt"

# Create the client instance
es = Elasticsearch(
    hosts = HOST,
    ca_certs = CA_CERTS,
    http_auth = (ELASTIC_USER, ELASTIC_PASSWORD),
)
es.info()
```

```python
# 2. create an index
index_name = "my_index"
type_name = "_doc"

if es.indices.exists(index_name):
    es.indices.delete(index_name)

es.indices.create(index_name)
es.cluster.health(wait_for_status="yellow")
```

```python
# 3. mapping
es.indices.put_mapping(
    index = index_name, 
    body = {
        "properties": {
            "uuid": {"type": "keyword"},
            "title": {
                "type": "text", 
                "term_vector": "with_positions_offsets"
            },
            "parsedtext": {
                "type": "text", 
                "term_vector": "with_positions_offsets"
            },
            "nested": {
                "type": "nested",
                "properties": {
                    "num": {"type": "integer"},
                    "name": {"type": "keyword"},
                    "value": {"type": "keyword"}
                }
            },
            "date": {"type": "date"},
            "position": {"type": "integer"},
            "name": {
                "type": "text", 
                "term_vector": "with_positions_offsets"
            }
        }
    }
)
```

```python
# 4. retrieve the mapping
import json

mappings = es.indices.get_mapping(index_name)
# prettyprint a JSON
print(json.dumps(mappings, indent=4, sort_keys=True))
#print(mappings)
```

```python
# check if the index is exists
es.indices.exists(index=index_name)
```

```python
# delete the index
es.indices.delete(index=index_name)
```

# II. Managing documents

```python
index_name = "my_index"
type_name = "_doc"
```

```python
# 1. index some documents
from datetime import datetime

es.index(
    index=index_name, 
    doc_type = type_name, 
    id=1, 
    body={
        "name": "Joe Tester", 
        "parsedtext": 
        "Joe Testere nice guy", 
        "uuid": "11111", 
        "position": 1, 
        "date": datetime(2018, 12, 8), 
        "join_field": {"name": "book"}
    }
)

es.index(
    index=index_name, 
    doc_type = type_name, 
    id="1.1", 
    body={
        "name": "data1", 
        "value": "value1", 
        "join_field": {"name": "metadata", "parent": "1"}}, 
    routing=1
)

es.index(
    index=index_name, 
    doc_type = type_name, 
    id=2, 
    body={
        "name": "Bill Baloney",  
        "parsedtext": "Bill Testere nice guy",
        "uuid": "22222", "position": 2, 
        "date": datetime(2018, 12, 8), 
        "join_field": {"name": "book"}
    }
)

es.index(
    index=index_name, 
    doc_type = type_name, 
    id="2.1", 
    body={
        "name": "data2", 
        "value": "value2", 
        "join_field": {"name": "metadata", "parent": "2"}
    }, 
    routing=2
)

es.index(
    index=index_name,
    doc_type = type_name, 
    id=3, 
    body={
        "name": "Bill Clinton", 
        "parsedtext": """Bill is not nice guy""", 
        "uuid": "33333", 
        "position": 3, "date": datetime(2018, 12, 8), 
        "join_field": {"name": "book"}
    }
)
```

```python
# 2. update a document
es.update(
    index=index_name, 
    #doc_type = type_name, 
    id=2, 
    body={"script": 'ctx._source.position += 1'}
)

document=es.get(index=index_name, id=2)
#print(document)
print(json.dumps(document, indent=4, sort_keys=True))
```

```python
# 3. delete a document
es.delete(
    index=index_name, 
    doc_type = type_name, 
    id=3)
```

```python
# 5. bulk insert some documents
from elasticsearch.helpers import bulk
bulk(es, [
    {
        "_index":index_name, 
        "_id":"1", 
        "source":{
            "name": "Joe Tester", 
            "parsedtext": 
            "Joe Testere nice guy", 
            "uuid": "11111", 
            "position": 1,
            "date": datetime(2018, 12, 8)}
    },

    {
        "_index": index_name, 
        "_id": "1",
        "source": {
            "name": "Bill Baloney", 
            "parsedtext": "Bill Testere nice guy", 
            "uuid": "22222", 
            "position": 2,
            "date": datetime(2018, 12, 8)
        }
    }
])
```

# III. Executing a standard search

```python
# 1. execute a search with a match_all query and print the results
from pprint import pprint

results = es.search(
    index = index_name,  
    doc_type = type_name, 
    body = {
        "query": {
            "match_all": {}
        }
    }
)
pprint(results)
```

```python
# 2. execute a search with a term query and print the results
results = es.search(
    index = index_name, 
    doc_type = type_name, 
    body = {
        "query": {
            "term": {
                "name": {
                    "boost": 3.0, 
                    "value": "joe"
                }
            }
        }
    }
)
pprint(results)
```

```python
# 3. execute a search with a bool filter query and print the results
results = es.search(
    index = index_name, 
    doc_type = type_name, 
    body = {
        "query": {
            "bool": {
                "filter": {
                    "bool": {
                        "should": [
                            {"term": {"position": 1}},
                            {"term": {"position": 2}}
                        ]
                    }
                }
            }
        }
    }
)
pprint(results)
```

# IV. Executing a search with aggregations

```python
# 1. execute a search with a terms aggregation
results = es.search(
    index = index_name, 
    doc_type = type_name, 
    body = {
        "size":0,
        "aggs": {
            "pterms": {
                "terms": {
                    "field": "name", 
                    "size": 3
                }
            }
        }
    }
)
pprint(results)
```

```python
# 2. execute a search with a date histogram aggregation
results = es.search(
    index = index_name, 
    doc_type = type_name, 
    body = {
        "size":0,
        "aggs": {
            "date_histo": {
                "date_histogram": {
                    "field": "date", 
                    "interval": "month"
                }
            }
        }
    }
)
pprint(results)
```
